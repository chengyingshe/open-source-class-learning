## torchvision

```python
import torchvision.models as models

resnet18 = models.resnet18(pretrained=True)
resnet18
```

```python
import torch.nn as nn
from collections import OrderedDict

in_dim, hidden_dim, out_dim = resnet18.fc.in_features, 256, 10
fc = nn.Sequential(OrderedDict({
    'fc1': nn.Linear(in_dim, hidden_dim),
    'relu': nn.ReLU(),
    'fc2': nn.Linear(hidden_dim, out_dim),
    'softmax': nn.Softmax()
}))
resnet18.fc = fc
resnet18
```

## timm

```python
# list all available models
import timm

avail_models = timm.list_models()
len(avail_models)
```

```python
# search model with filter
timm.list_models('*densenet*')
```

```python
# create pretrained model
model = timm.create_model('densenet121', pretrained=True)
model.classifier
```

```python
# modify the output layer(num_classes) of the pretrained model
model = timm.create_model('densenet121', pretrained=True, num_classes=10)
model.classifier
```

```python
# inference
import torch

sample = torch.randn(1, 3, 224, 224)
model(sample)
```

```python
# save model
torch.save(model.state_dict(), 'models/densenet121_out10.pt')
```

```python
# load model
new_model = timm.create_model('densenet121', pretrained=False, num_classes=10)
new_model.load_state_dict(torch.load('models/densenet121_out10.pt'))
new_model
```
