## map和zip函数

```python
list(map(lambda x: x**2, [1, 2, 3, 4, 5]))  # 将前面的函数应用到后面列表中的每一个值（返回的是一个map对象）
```

```python
l1 = list('abc')
l2 = list('123')
l3 = list('xyz')
list(zip(l1, l2, l3))  # 将两个列表中的值一一对应起来（返回的是一个zip对象）
```
